```python
import pandas as pd
titanic = pd.read_csv('train.csv')
titanic
```

```python
titanic.head()
titanic.tail()
```

```python
# Drop duplicates
titanic.drop_duplicates(inplace=True)
titanic.reset_index(inplace=True,drop=True)
titanic
```

```python
# Checking for null values
null_counts = titanic.isnull().sum()
if null_counts.sum()>0:
    print(null_counts[null_counts>0])
```

```python
titanic = titanic[titanic['Embarked'].notna()]
titanic['Age'].fillna(titanic['Age'].median(), inplace=True)
#titanic = titanic.drop('Cabin', axis =1)
null_counts = titanic.isnull().sum()
if null_counts.sum()>0:
    print(null_counts[null_counts>0])
```

```python
titanic
```

```python
titanic.dtypes
```

```python
titanic['Age']=titanic['Age'].astype('int64')
```

```python
import seaborn as sns
sns.boxplot(titanic['Age'])
```

```python
import numpy as np
z_scores = np.abs((titanic[['Age','SibSp']]-titanic[['Age','SibSp']].mean())/titanic[['Age','SibSp']].std())
outliers_threshold=3
outliers = titanic[(z_scores['Age']>outliers_threshold)|(z_scores['SibSp']>outliers_threshold)]
print(outliers)
```

```python
age_threshold = titanic['Age'].quantile(0.95)
SibSp_threshold = titanic['SibSp'].quantile(0.95)
titanic['Age']=np.where(titanic['Age']>age_threshold,age_threshold,titanic['Age'])
titanic['Sibsp']=np.where(titanic['SibSp']>SibSp_threshold,SibSp_threshold,titanic['SibSp'])
titanic
```

```python
titanic.drop(['PassengerId','Name','Ticket'],axis=1,inplace=True)
```
